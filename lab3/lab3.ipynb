````python
data_path = "zombie/raw/zombies-survey.csv"
````

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import pearsonr, spearmanr
import seaborn as sns
````

````python
def calculate_p_values(df, method):
    df = df.dropna()._get_numeric_data()
    df_columns = pd.DataFrame(columns=df.columns)
    p_values = df_columns.transpose().join(df_columns, how='outer')
    for r in df.columns:
        for c in df.columns:
            if method == 'pearson':
                p_values[r][c] = str(pearsonr(df[r], df[c])[1])
            else:
                p_values[r][c] = str(spearmanr(df[r], df[c])[1])
    return p_values
````

````python
df_zombies_survey = pd.read_csv(data_path)
df_zombies_survey
````

Como podemos ver acima, nosso conjunto de dados tem informações numéricas e informações textuais. Podemos separa as informações numéricas em um outro dataframe para poder manipulá-las de forma mais direta e extrair estatísticas.

````python
df_numeric_data = df_zombies_survey[['height', 'weight', 'gbc', 'gs', 'pbc', 'sp']]
df_numeric_data
````

A funções do pandas chamada ```describe()``` nos fornece algumas informnações estatísticas sobre cada variável no conjunto de dados como contagem de entradas não vazias por coluna, média, mediana, desvio padrão, valor mínimo, 25, 50 e 75 percentis e valor máximo.

Isso nos ajuda a ter uma noção da tendência central e a dispersão do conjunto de dados.

````python
df_numeric_data.describe()
````

Podemos também gerar um scatter plot de cada variável para analizar visualmente como os dados estão distribuidos no espaço.

````python
pd.plotting.scatter_matrix(df_numeric_data, figsize=(20,20), marker='o', grid=True, range_padding=1, alpha=0.5)
````

Pela diagonal principal, podemos observar que algumas variáveis possuvem distribuição normal, como as `sp, weight, height`, e outras não possuem distrubuição normal como `pbc, gs, gbc`

Ainda da análise visual conseguimos observar que algumas variáveis aparentam ter forte correlação positiva, como `weight e height`, `gs e gbc`, `gs e pbc` e `pbc e gbc`, enquanto outras não aparentem possuir correlação, como as variáveis `weight e height` em relação as demais variáveis.


## Correlação de Pearson

Podemos obter os valores numéricos de correlação entre cada par de variáveis para confirmar ou não essas observações.
Para esse primeiro momento, utilizaremos a correlação de Pearson, mas devemos nos atentar que ela pode não ser a mais adequada para todas as variávels, pois, como apontado anteriormente, algumas variáveis não possuem distribuição normal, ou seja, estão distribuídas segundo uma gaussiana.

````python
R = df_numeric_data.corr(method='pearson')
R
````

A partir da tabela mostrada acima, podemos confirmar que as variáveis que aparentavam visualmente possuírem forte correlação positiva realmente a possuem, com valores de correlação todos acima de 0.8, e podemos confirmar que as variáveis `weight e height` realmente não possuem correlação com as demais variáveis, com valores de R muito próximos de 0. Além disso, podemos ver que algumas variáveis possuem correlação negativa fraca, como `sp e gbc` e `sp e gs`.

Precisamos também medir o grau de certeza dos valores de correlação. Para isso calculamos os P-valores associados a essas correlações.

````python
p_values = calculate_p_values(df_numeric_data, 'pearson')
p_values
````

O p-valor representa o nível de certeza das medidas de correlação obtidas. Caso `p-valor < 0.001` temos uma baixa probabilidade de que a medida não está correta e portanto temos um alto nível de certeza. Podemos observar que esse é o caso dos pares `weight e height`, `gs e gbc`, `pbc e gs` e `pbc e gbc`, que apresentavam uma forte correlação positiva na análise anterior. Além disso, também se encaixam nessa faixa de p-valor os dois pares com correlação negativa fraca, que eram `sp e gs` e `sp e gbc`.

Caso `p-valor < 0.05` temos um nível de certeza moderado, porém não temos nenhum par de variáveis nessa faixa de valor. O mesmo ocorre para `p-valor < 0.1`, onde temos um nível de certeza fraco, mas também nenhum par nessa faixa de valor.

Caso `p-valor > 0.1`, não há certeza aceitável no resultado e esse é o caso dos pares de variáveis `weight e height` em relação a todas as demais variávels, e também do par `sp e pbc`.

## Correlação de Spearman

Como apontado anteriormente, algumas das nossas variáveis aparentam apresentar uma distribuição não normal. Vamos utilizar então a correlação de Spearman, que é o equivalente não-paramétrico da correlação de Pearson para novamente observar como se comportam as correlações.

````python
R = df_numeric_data.corr(method='spearman')
R
````

A partir da tabela mostrada acima, podemos observar que os mesmos pares de variáveis continuam com forte correlação positiva, são eles `weight e height`, `gs e gbc`, `gs e pbc` e `pbc e gbc`, todos com valores de correlação acima de 0.8. Também seguimos com as variáveis `weight e height` sem correlação com as demais variáveis, com valores de R muito próximos de 0.

Além disso, podemos ver que as variáveis `sp e gs` possuem correlação negativa fraca (R > 0.3).

Precisamos também medir o grau de certeza dos valores de correlação. Para isso calculamos os P-valores associados a essas correlações.

````python
p_values = calculate_p_values(df_numeric_data, 'spearman')
p_values
````

Podemos observar que para o caso `p < 0.001`, onde temos um alto nível de certeza do valor de correlação, temos os pares `weight e height`, `gs e gbc`, `pbc e gbc` e `gs e pbc` com alta correlação positiva e o par `sp e gs` com fraca correlação negativa. Mesmos pares de observados para a correlação de Pearson.
