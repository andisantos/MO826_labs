```python
import sklearn
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr

from sklearn.linear_model import LogisticRegressionCV, LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
def calculate_p_values(df, method):
    df = df.dropna()._get_numeric_data()
    df_columns = pd.DataFrame(columns=df.columns)
    p_values = df_columns.transpose().join(df_columns, how='outer')
    for r in df.columns:
        for c in df.columns:
            if method == 'pearson':
                p_values[r][c] = str(pearsonr(df[r], df[c])[1])
            else:
                p_values[r][c] = str(spearmanr(df[r], df[c])[1])
    return p_values
```

<h2><center>Análise 2: Regressão Linear</center></h2>


**Tarefa**: Considerando a base Zombie Meals (zombie-meals.csv) que lista refeições típicas de zumbis e sua caracterização em termos de proteínas, calorias, sódio, etc.  A última coluna (rating) representa a nota média da avaliação feita pelos zumbis para aquela refeição.


Realize uma regressão linear e analise a influência destas características (proteínas, calorias, sódio, etc.) dos alimentos com as avaliações (ratings) feitas pelos zumbis. Analise as características que mais influenciam na avaliação (se houver) e monte um modelo de regressão (com uma ou múltiplas variáveis independentes) que você acha que melhor prediz o rating. Considere aspectos como influência dos parâmetros, correlação, R2 e overfitting.

```python
df_zombie_meals.describe()
```

```python
R_pearson = df_zombie_meals.corr(method = 'pearson')
R_pearson
```

```python
pvalores_pearson = calculate_p_values(df_zombie_meals, method='pearson')
pvalores_pearson
```

```python
R_spearman = df_zombie_meals.corr(method='spearman')
R_spearman
```

```python
pvalores_spearman = calculate_p_values(df_zombie_meals, method = 'spearman')
pvalores_spearman
```

```python
X = df_zombie_meals[['calories', 'fiber', 'sugars']]
y = df_zombie_meals['rating']
```

```python
linear_regression = LinearRegression(normalize=True)
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.25,
                                                    random_state = 42,
                                                    shuffle = True)
```

```python
print("Tamanho do conjunto de treino = {} amostras\nTamanho do conjunto de teste = {} amostras".format(
    X_train.size,
    X_test.size))
```

```python
linear_regression = linear_regression.fit(X_train, y_train)
r_square = linear_regression.score(X_test, y_test)
print("Intercept (a):", linear_regression.intercept_)
print('Coeficientes:', linear_regression.coef_)

print("R2:", r_square)
```

<h2><center>Análise 3: Regressão Logística</center></h2>

**Tarefa**: Considerando a base Zombies Survey (zombies-survey.csv), realize uma regressão logistica para predizer o gênero dos zumbis a partir das outras features. Analise as features que mais influenciam na predição (se houver) e monte um modelo de regressão (com uma ou múltiplas variáveis independentes) que você acha que melhor prediz o gênero.
```python
df_zombie_survey['gender_id'] = np.where(df_zombie_survey['gender']=='male', 1, 0)
df_zombie_survey.head()
```

Vamos primeiramente analisar as variáveis do dataset zombie-survey.csv. As análises visam identificar possíveis comportamentos de cada variaǘel em relação às demais.

Para isso vamos computar os valores de correlação de Pearson e Spearman.

```python
df_zombie_survey.describe()
```

```python
df_zombie_survey_numeric_data = df_zombie_survey[['height','weight', 'gbc', 'gs', 'pbc', 'sp', 'gender_id']]
```

```python
# correlações de pearson
R_pearson = df_zombie_survey_numeric_data.corr(method='pearson')
R_pearson
```

```python
# P-valor para as variaves utilizando correlação de pearson
pvalores_pearson = calculate_p_values(df_zombie_survey_numeric_data, method='pearson')
pvalores_pearson
```

```python
# correlações de spearman
R_spearman = df_zombie_survey_numeric_data.corr(method='spearman')
R_spearman
```

```python
# P-valor para as variaves utilizando correlação de spearman
pvalores_spearman = calculate_p_values(df_zombie_survey_numeric_data, method='spearman')
pvalores_spearman
```

Podemos observar que para a variável **gender_id**, que apresenta o valor 0 (zero) quando gender = female e 1 (um) quando gender = male, apresenta forte correlação com as variáveis height e weight com alto nível de certeza, ou seja, p-valor < 0.001 tanto para correlção de Pearson quanto para Spearman.

Além disso, as demais variáveis apresentam baixa correlação com a variável **gender_id**, com valores muito próximos de 0 e alto p-valor, indicando que estas variáveis não podem nos auxiliar na predição da variável gender_id.

Vamos utilizar variáveis height e weight para a regressão logística.

<h2><center>Construção e treinamento do modelo</center></h2>

```python
regressor = LogisticRegressionCV(random_state = 42, cv = 5)
```

```python
X = df_zombie_survey_numeric_data[['height', 'weight']]
y = df_zombie_survey_numeric_data['gender_id']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.30,
                                                    random_state = 42,
                                                    shuffle = True)
```

```python
print("Tamanho do conjunto de treino = {} amostras\nTamanho do conjunto de teste = {} amostras".format(
    X_train.size,
    X_test.size))
```

```python
regressor = regressor.fit(X_train, y_train)
```

<center><h2> Avaliação do modelo de Regressão Logística</center></h2>

```python
test_predictions = regressor.predict(X_test)
```

```python
report = classification_report(y_test, test_predictions, target_names=['female', 'male'], output_dict=True)
```

```python
report_df = pd.DataFrame(report)
report_df
```

```python
c_matrix = confusion_matrix(y_test, test_predictions)

ticklabels = ['female', 'male']
ax = plt.axes()
sns.heatmap((c_matrix/sum(c_matrix)), annot=True,
            xticklabels=ticklabels, 
            yticklabels=ticklabels,
            fmt='.02%',
            cmap=sns.light_palette("seagreen", as_cmap=True),
            linewidths=0.2, ax = ax)

ax.set_title('Matriz de Confusão')
plt.xlabel('Gênero Predito', fontsize = 15)
plt.ylabel('Gênero Real', fontsize = 15)
plt.show()
```
